```python
with open("Data/000/Trajectory/20081024020959.plt") as f:
    trajectory = f.read().splitlines()
    trajectory = trajectory[6:]
```

```python
for i in range(len(trajectory)):
    data = trajectory[i].split(',')[:2]
    data = list(map(float, data))
    trajectory[i] = data
```

# Data Visualization

https://jakevdp.github.io/PythonDataScienceHandbook/04.13-geographic-data-with-basemap.html

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution=None,
            width=4E6, height=4E6, 
            lat_0=40.008304, lon_0=116.319876,)
m.etopo(scale=0.5, alpha=0.5)

# Map (long, lat) to (x, y) for plotting
x, y = m(116.320214, 40.007694)
plt.plot(x, y, 'ok', markersize=5)
```
